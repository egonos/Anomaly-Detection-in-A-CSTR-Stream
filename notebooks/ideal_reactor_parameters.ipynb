```python
import sys
import os

sys.path.append(os.path.abspath("../src"))

import seaborn as sns
import matplotlib.pyplot as plt
from utils import generate_normal_samples
```

```python
initial_samples = generate_normal_samples(n_samples=10000, input_noise_scale=0.03)

temp_values = initial_samples.temperature.values
Ca0_values = initial_samples.Ca0.values
Q_values = initial_samples.Q.values
Ca_values = initial_samples["Ca"].values
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
sns.histplot(temp_values, bins=50, kde=True, ax=ax[0][0])
sns.histplot(Ca0_values, bins=50, kde=True, ax=ax[0][1])
sns.histplot(Q_values, bins=50, kde=True, ax=ax[1][0])
sns.histplot(Ca_values, bins=50, kde=True, ax=ax[1][1])
ax[0][0].set_title("Temperature")
ax[0][1].set_title("Ca0")
ax[1][0].set_title("Q")
ax[1][1].set_title("Ca")
plt.tight_layout()
fig.show()
```
